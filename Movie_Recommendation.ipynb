'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'

We Have To Convert The Above Format Into Below Format


['Action','Adventure','Fantasy','SciFi']

```python
#Helper Functions (Converter)

def convert(object):
  L=[]
  for i in object:
    L.append(i['name'])
  return L


convert([{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}])
```

OR

If We Pass The List In The Convert Function, An Error Will Appear Which Shows "String Indices Must Be Integers" Then We Can Use AST Module To Convert String Into Integer.

```python
import ast
def convert(object):
    L=[]
    for i in ast.literal_eval(object):
        L.append(i['name'])
    return L
```

## Removing Spaces

```python
dataset_movies['genres']=dataset_movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
dataset_movies['keywords']=dataset_movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
dataset_movies['cast']=dataset_movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
dataset_movies['crew']=dataset_movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

## Merging All Columns(Overview,Genres,Keywords,Cast,Crew) Into One (Tags)

```python
dataset_movies['tags']=dataset_movies['overview'] + dataset_movies['genres'] + dataset_movies['keywords'] + dataset_movies['cast'] + dataset_movies['crew']
```

```python
dataset_movies
```

```python
final_dataset = dataset_movies[['id','title','tags']]
final_dataset
```

## Join And Lower Case In Tags Column

```python
final_dataset['tags'] = final_dataset['tags'].apply(lambda x:" ".join(x))
```

```python
final_dataset
```

```python
final_dataset['tags'] = final_dataset['tags'].apply(lambda x:x.lower())
```

# Vectorization (Bag of Word)

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
vectors = cv.fit_transform(final_dataset['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
# It Will Recommend Us 5 Similiar Movies
# Firstly We Will Find Out The Index of The Movies
# Then Moves To Similarity Index Which Is Actually In Numpy Array Form
# Now We Will Sort The Distance of Movie With Respect To Other Movies Similarity Distance.
# Fetch Top 5 Movies

# Firstly Fetch Index of The Each Movie
final_dataset['title'] == 'Avatar' # Output Will Be True Or False

final_dataset[final_dataset['title']== 'Avatar'].index[0] # Output Will Be Its Index Number
```

```python
def recommend(movie):
    movie_index = final_dataset[final_dataset['title']== movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x:x[1])[1:6]

    for i in movies_list:
        print(final_dataset.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
